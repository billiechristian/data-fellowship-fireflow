```python
import pandas as pd
import random
```

```python
cust = pd.read_csv("customer.csv")
```

```python
cust.drop("Unnamed: 0", axis=1).to_csv("customer.csv", index=False)
```

```python
cust['sex'] = [random.choice(['male', 'female']) for i in range(len(cust))]
```

```python
cust["city"] = [random.choice(['Houston', 'New York', 'Phoenix', 'Chicago', 'Los Angeles']) for i in range(len(cust))]
```

```python
cust = cust[['id', 'age', 'sex', 'job_id', 'marital', 'education_id', 'default', 'housing',
       'loan', 'city']]
```

```python
cust.to_csv("customer.csv")
```

```python
job = pd.read_csv("job.csv")
job
```

```python
job["average_age"] = job.average_age.apply(lambda x: round(x,2))
```
